```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
len(soup.find_all('div', class_='sm-product has-tag has-features has-actions'))
```

```python
import numpy as np
import pandas as pd
```

```python
names = []
prices = []
ratings = []
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []

for i in soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'}):
    try:
        names.append(i.find('h2').text)
    except:
        names.append(np.nan)
    try:
        prices.append(i.find('span',{'class':'price'}).text)
    except:
        price.append(np.nan)
    try:
        ratings.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
    except:
        ratings.append(np.nan)

    x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    try:
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
    try:
        ram.append(x[2].text)
    except:
        ram.append(np.nan)
    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
    try:
        camera.append(x[5].text)
    except:
        camera.append(np.nan)
    try:
        card.append(x[6].text)
    except:
        card.append(np.nan)
    try:
        os.append(x[7].text)
    except:
        os.append(np.nan)
```

```python
df = pd.DataFrame({
    'model':names,
    'price':prices,
    'rating':ratings,
    'sim':sim,
    'processor':processor,
    'ram':ram,
    'battery':battery,
    'display':display,
    'camera':camera,
    'card':card,
    'os':os
})
```

```python
# Clean the 'price' column and convert to integer
df['price'] = df['price'].str.replace('₹', '').str.replace(',', '').astype(int)
```

```python
# Remove content inside parentheses and parentheses themselves from 'model' column
df['model'] = df['model'].str.replace(r"\(.*\)", "", regex=True).str.strip()
```

```python
data = df[df['price']>8000]
```

```python
data['processor_name'] = data['processor'].str.split(',').str.get(0)
data['processor_core'] = data['processor'].str.split(',').str.get(1)
data['processor_speed(GHz)'] = data['processor'].str.split(',').str.get(2).str.replace("GHz Processor", "").str.strip().astype('float')
```

```python
data.head(1)
```

```python
data['Ram'] = data['ram'].str.split(',').str.get(0).str.extract('(\d+)').astype('int')
```

```python
data['Rom'] = data['ram'].str.split(',').str.get(1).fillna('0').str.extract('(\d+)').astype('int')
```

```python
data.head(1)
```

```python
# Drop specified columns
columns_to_drop = ['processor', 'card', 'os', 'ram']
data.drop(columns=columns_to_drop, inplace=True)
```

```python
data.head(1)
```

```python
data['battery_mah'] = data['battery'].str.extract(r'(\d+)\s?mAh').fillna(0).astype(int)
```

```python
data.drop(columns={'battery'}, inplace=True)
```

```python
data
```

```python
data['rating'] = data['rating'].astype('float')
```

```python
data['rating'].fillna(67, inplace=True)
```

```python
data.nlargest(5, 'rating')
```

```python
data['refresh_rate(Hz)'] = data['display'].str.split(',').str.get(2).str.extract(r'(\d+) Hz').fillna(0).astype(int)
data['display_size(inces)'] = data['display'].str.split(',').str.get(0).str.extract(r'(\d+\.?\d*)').fillna('0').astype(float)
```

```python
columns_to_drop = ['display']
data.drop(columns=columns_to_drop, inplace=True)
```

```python
data
```

```python
data['company'] = data['model'].str.split(" ").str.get(0)
```
